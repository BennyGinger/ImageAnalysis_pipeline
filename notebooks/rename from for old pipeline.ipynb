```python
# rename according to old pipeline
folder = 'E:\Fabian\FibroblastPaper\Version2'

import os
import re 
for root, dirs, files in os.walk(folder, topdown=False):
   for name in dirs:
      print(os.path.join(root, name))
      if name == 'Masks_IoU_Track':
         os.rename(os.path.join(root, name),os.path.join(root, 'Masks_stitched'))
      elif name == 'Masks_Cellpose':
         os.rename(os.path.join(root, name),os.path.join(root, 'Masks_CP'))
      else:
         seriesre = re.search('s\d+', name)
         if not seriesre == None:
            series = seriesre.group(0)
            splitname = name.split('_')
            if series == splitname[-1]: 
               newname = splitname[-1]+'_'+'_'.join(splitname[:-1])
               os.rename(os.path.join(root, name),os.path.join(root, newname))
```
